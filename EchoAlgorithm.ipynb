```python
# Fill train_echo and test_echo lists with approx. equal number of echo files

train_echo = []
test_echo = []

if lang == 'ENGESV':
    for file in wav_files:
        for verse in ['10', '11', '12', '13', '14', '15', '16', '17', '18']:
            if file[67:69] == '03' and file[93:95] == verse:
                train_echo.append(file)
        for verse in ['6_', '7_', '11', '13']:
            if file[67:69] == '10' and file[93:95] == verse:
                train_echo.append(file)
        if file[67:69] == '01' and file[93:95] == '17':
            test_echo.append(file)
        if file[67:69] == '02' and file[93:95] == '24':
            test_echo.append(file)
        for verse in ['3_', '7_', '8_', '17', '18']:
            if file[67:69] == '04' and file[93:95] == verse:
                test_echo.append(file)
        if file[67:69] == '08' and file[93:95] == '36':
            test_echo.append(file)
        for verse in ['15', '16', '20', '21']:
            if file[67:69] == '10' and file[93:95] == verse:
                test_echo.append(file)
elif lang == 'INDWBT':
    for file in wav_files:
        for verse in ['10', '11', '12', '13', '14', '15', '16', '17', '18']:
            if file[67:69] == '03' and file[93:95] == verse:
                train_echo.append(file)
        if file[67:69] == '01' and file[93:95] == '17':
            test_echo.append(file)
        if file[67:69] == '02' and file[93:95] == '24':
            test_echo.append(file)
        for verse in ['3_', '7_', '8_', '17', '18']:
            if file[67:69] == '04' and file[93:95] == verse:
                test_echo.append(file)
        if file[67:69] == '08' and file[93:95] == '36':
            test_echo.append(file)
        if file[67:69] == '10' and file[93:95] == '18':
            test_echo.append(file)
elif lang == 'CMNUN1':
    for file in wav_files:
        for verse in ['10', '11', '12', '13', '14', '15', '16', '17', '18']:
            if file[67:69] == '03' and file[93:95] == verse:
                train_echo.append(file)
        if file[67:69] == '01' and file[93:95] == '17':
            test_echo.append(file)
        for verse in ['3_', '7_', '8_', '17', '18']:
            if file[67:69] == '04' and file[93:95] == verse:
                test_echo.append(file)
        if file[67:69] == '08' and file[93:95] == '36':
            test_echo.append(file)
        for verse in ['5_', '20']:
            if file[67:69] == '10' and file[93:95] == verse:
                test_echo.append(file)
print(len(train_echo))
print(len(test_echo))
```

```python
# Convert wav files to Parselmouth sounds

train_sounds = []
test_echo_sounds = []
all_sounds = []

for wav_file in train_echo:
    sound = parselmouth.Sound(wav_file)
    sound.name = wav_file
    train_sounds.append(sound)
for wav_file in test_echo:
    sound = parselmouth.Sound(wav_file)
    sound.name = wav_file
    test_echo_sounds.append(sound)
for wav_file in wav_files:
    sound = parselmouth.Sound(wav_file)
    sound.name = wav_file
    all_sounds.append(sound)
print(train_sounds[0].name)
print(test_echo_sounds[0].name)
```

```python
# Get avg hnr and shimmer for train echo set

count = 0
avg_echo_hnr= 0
avg_echo_shimmer = 0
total_hnr = 0
total_shimmer = 0

for sound in train_sounds:
    harmonicity = call(sound, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)
    
    f0min = 75
    f0max = 500
    pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)
    shimmer =  call([sound, pointProcess], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    
    print(sound.name[67:69] + ":" + sound.name[93:95] + " " + str(shimmer))
    
    total_hnr += hnr
    total_shimmer += shimmer
    count += 1
avg_echo_hnr = total_hnr / count
avg_echo_shimmer = total_shimmer / count
print(avg_echo_hnr, avg_echo_shimmer)
```

```python
# avg HNR and shimmer for non echo files
# these values are from the Speech Features Observations sheet

if lang == 'ENGESV':
    hnr_nonecho = 10.216
    shimmer_nonecho = 0.097
# change these later
if lang == 'CMNUN1':
    hnr_nonecho = 10.77
    shimmer_nonecho = 0.099
if lang == 'INDWBT':
    hnr_nonecho = 15.645
    shimmer_nonecho = 0.08
print(hnr_nonecho)
print(shimmer_nonecho)
```

```python
import math

hnr_echo_files = []
shimmer_echo_files = []
echo_files = []
nonecho_files = []

for sound in all_sounds:
    harmonicity = call(sound, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)
    
    f0min = 75
    f0max = 500
    pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)
    shimmer =  call([sound, pointProcess], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    
    # the threshold number (i.e. 1/3) comes from the number multiplied at the end
    # threshold determines how strict the echo algorithm should be
    hnr_threshold = abs(avg_echo_hnr - hnr_nonecho) *1 / 3
    shimmer_threshold = abs(avg_echo_shimmer - shimmer_nonecho) *1 / 3
    
    if hnr < (avg_echo_hnr+hnr_threshold):
        if shimmer > (avg_echo_shimmer - shimmer_threshold): # closer to echo, by hnr and shimmer
            echo_files.append(sound)
        else:
            hnr_echo_files.append(sound) # only echo by hnr threshold
    elif shimmer > (avg_echo_shimmer - shimmer_threshold): # only echo by shimmer threshold
        shimmer_echo_files.append(sound)
    else:
        nonecho_files.append(sound)
```

```python
# print files names by algorithm's classification
# ignore the shimmer only and hnr only echo

print("echo", len(echo_files))
for file in echo_files:
    print(file.name)
print("\nshimmer only echo", len(shimmer_echo_files))
for file in shimmer_echo_files:
    print(file.name)
print("\nhnr only echo", len(hnr_echo_files))
for file in hnr_echo_files:
    print(file.name)
print("\nnon echo", len(nonecho_files))
for file in nonecho_files:
    print(file.name)    
```

```python
# put original data (train + test files) with correct labels in a dataframe

import pandas as pd
data = []

all_echo_sounds = train_sounds + test_echo_sounds
print(len(all_echo_sounds))

for sound in all_sounds:
    for echo_sound in all_echo_sounds:
        if sound.name[93:95] == '0_':
            break
        if sound.name[61:-4] == echo_sound.name[61:-4]:
            data.append([sound.name[67:69] + ":" + sound.name[93:95], sound.name, 'E'])
            break
    else:
        data.append([sound.name[67:69] + ":" + sound.name[93:95], sound.name, 'NE'])
    
df = pd.DataFrame(data, columns=['verse', 'full_name', 'label'])
df.info()
```

```python
# results of echo algorithm

correct_echo = 0
correct_nonecho = 0
false_pos = 0 # classified as echo but actually nonecho
false_neg = 0 # classified as nonecho but actually echo

all_nonecho_files = shimmer_echo_files + hnr_echo_files + nonecho_files

for index, row in df.iterrows():
    sound_name = row['full_name']
    label = row['label']
    for sound in echo_files:
        if sound_name == sound.name:
            if label == 'E':
                correct_echo += 1
            elif label == 'NE':
                print("fP", sound_name)
                false_pos += 1
for index, row in df.iterrows():
    sound_name = row['full_name']
    label = row['label']
    for sound in all_nonecho_files:
        if sound_name == sound.name:
            if label == 'NE':
                correct_nonecho += 1
            elif label == 'E':
                print("fN", sound_name)
                false_neg += 1

print("correct echo:", correct_echo, "correct nonecho:", correct_nonecho, "false pos:", false_pos, "false neg:", false_neg)
```
